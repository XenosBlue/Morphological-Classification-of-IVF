```python
def save_img(img,name):
  cv2.imwrite(name, img)
```

```python
def preprocess(img):

  img  = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  img = cv2.GaussianBlur(img,(5,5),0)
  blur = cv2.GaussianBlur(img,(5,5),0)
  aggregate =[]
  step = 1
  till = 101
  for i in range(0,till,step):
    high = i+step
    low  = i
    canny_img = cv2.Canny(np.uint8(blur),low,high)
    aggregate.append(canny_img)
  aggregate = np.stack(aggregate, axis=0)
  accumulate = np.mean(aggregate,axis=0).astype('float32')
  del aggregate
  #accumulate = ( accumulate - np.amin(accumulate) ) / ( np.amax(accumulate) - np.amin(accumulate) )

  return np.stack([img,accumulate,blur],axis=2)
```

```python
dir = 'D3_T_N'
counter = 0
for filename in os.listdir(dir):
    if filename.endswith("png"):
        counter += 1
        img = cv2.imread(dir+'/'+filename, cv2.IMREAD_COLOR)
        save_img(preprocess(img),'meta'+dir+'/'+filename)
        print('meta'+dir+'/'+filename)
print(counter)
```

```python
dir = 'D3_T_Y'
counter = 0
for filename in os.listdir(dir):
    if filename.endswith("png"):
        counter += 1
        img = cv2.imread(dir+'/'+filename, cv2.IMREAD_COLOR)
        save_img(preprocess(img),'meta'+dir+'/'+filename)
        print('meta'+dir+'/'+filename)
print(counter)
```
